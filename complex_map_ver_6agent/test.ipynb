```python
import numpy as np

# サンプルの四次元配列を作成（例）
new_theta2 = np.random.rand(2, 3, 4, 5)  # 2x3x4x5のランダム配列
new_theta2[0, 0, 0, 0] = np.nan  # np.nanを含むようにする

# 最大値を計算し、新しい配列を作成する関数
def normalize_4d_array(array):
    # 配列内の最大値（np.nanを除外）を取得
    max_value = np.nanmax(array)
    
    # 最大値を引いた新しい配列を作成
    result = array - max_value
    
    # np.nanの位置を保持
    result[np.isnan(array)] = np.nan
    return result

# 新しい配列を作成
result_array = normalize_4d_array(new_theta2)

print("元の配列:\n", new_theta2)
print("\n新しい配列:\n", result_array)
```

```python
new_theta2 = np.random.rand(2, 3, 4, 5)  # 2x3x4x5のランダム配列
new_theta2[0, 0, 0, 0] = np.nan  # np.nanを含むようにする
new_theta2[1, 0, 0, 0] = np.nan 
print(np.nanmax(new_theta2, axis=(2, 3), keepdims=True))
max_theta = np.nanmax(new_theta2, axis=(2, 3), keepdims=True)
new_theta2 = new_theta2 - max_theta  # 数値安定化した theta
print("元の配列:\n", new_theta2)
```
